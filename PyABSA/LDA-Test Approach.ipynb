```python
import pandas as pd
```

```python
df=pd.read_csv('data.csv')
colAttributes = ['content', 'date', 'variant', 'author', 'product']
df=df[colAttributes]
df.dropna(inplace=True)
```

```python
df
```

```python
#Remove the rows which have no content
df=df[df['content'] != '']
df
```

```python
#Remove the rows that have content which are not in English

import langid
df['lang'] = df['content'].apply(lambda x: langid.classify(x)[0])
df=df[df['lang'] == 'en']
df
```

```python
#Generate topics for each content and create a dataframe to store the review and cooresponding topic

from gensim import corpora, models
import gensim
import nltk
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import string
import re
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

stop = set(stopwords.words('english'))
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()

def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized

doc_clean = [clean(doc).split() for doc in df['content']]
dictionary = corpora.Dictionary(doc_clean)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
Lda = gensim.models.ldamodel.LdaModel
ldamodel = Lda(doc_term_matrix, num_topics=10, id2word = dictionary, passes=50)
topics = ldamodel.print_topics(num_topics=10, num_words=5)
topics
```

```python
#Create a dataframe to store the review and cooresponding topic
df['topic'] = df['content'].apply(lambda x: ldamodel.get_document_topics(dictionary.doc2bow(clean(x).split()))[0][0])
df
```

```python
#store topic id and topic name in a dictionary
topicDict = {}
for i in range(0,10):
    topicDict[i] = topics[i][1]

topicDict
```

```python
#Identify poisitive and negative reviews

from textblob import TextBlob
df['polarity'] = df['content'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['sentiment'] = df['polarity'].apply(lambda x: 'positive' if x > 0 else 'negative')
df
```

```python
from pyabsa import available_checkpoints
checkpoint_map = available_checkpoints()
```

```python
from pyabsa import ATEPCCheckpointManager

aspect_extractor = ATEPCCheckpointManager.get_aspect_extractor(checkpoint='english',
                                   auto_device=True  # False means load model on CPU
                                   )
```

```python
examples = [df['content'][3]]
inference_source = examples
atepc_result = aspect_extractor.extract_aspect(inference_source=inference_source,pred_sentiment=True,  save_result=False)
```
